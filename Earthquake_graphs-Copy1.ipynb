```python
#globe map plot
eq_lat,eq_lon=[],[]
magnitudes=[]
eq_ts=[]
with open('C:\\Users\\TORAL\\Downloads\\Earthquake.csv') as f:
    reader=csv.reader(f)
    next(reader) 
    for row in reader:
        eq_lat.append(float(row[1]))
        eq_lon.append(float(row[2]))
        magnitudes.append(float(row[4]))
        eq_ts.append(row[0])
def mk_color(magnitude):
    if magnitude<3.0:
        return('go')
    elif magnitude<4.5: 
        return('yo')
    else:
        return('ro')
plt.figure(figsize=(15,11))
my_map=Basemap(projection='robin',resolution='1',area_thresh=1000.0,lat_0=0,lon_0=0)
my_map.drawcostallines()
my_map.drawcountries()
my_map.fillcontinents(color='#AA96DA')
my_map.drawmapboundary()
my_map.drawmeridians(np.arrange(0,360,30))
my_map.drawparallels(np.arrange(-90,90,30))
mk_size=2.4
for lon,lat,mag in zip(eq_lon,eq_lat,magnitudes):
    x,y=my_map(lon,lat)
    msize=mag*mk_size
    marker_string=mk_color(mag)
    my_map.plot(x,y,marker_string,markersize=msize)
plt.title('Earthquakes')
plt.savefig('eq_data.png')
plt.show()
```
